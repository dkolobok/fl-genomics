```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/home/genxadmin/accuracy.log', sep=' ', skiprows=1, header=None)
data.columns = ['number', 'train', 'validation', 'test']
numbers = data['number'].unique()

train = []
validation = []
test = []
for number in numbers:
    subset = data.loc[data['number'] == number]
    
    train.append(subset['train'].sum() / len(subset))
    validation.append(subset['validation'].sum() / len(subset))
    test.append(subset['test'].sum() / len(subset))

plt.plot(numbers, train)
plt.plot(numbers, validation)
plt.plot(numbers, test)
plt.xticks(numbers)
plt.grid()
plt.show()
```
